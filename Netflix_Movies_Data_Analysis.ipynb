**Categoring Vote_Average Column**

*We divide the Vote_Average values into four categories — **popular, average, below_avg, and not_popular** — to better describe the data using the categorize_col() function shown above*

```python
def categorize_col(df, col, labels):

    edges = [df[col].describe()['min'],
            df[col].describe()['25%'],
            df[col].describe()['50%'],
            df[col].describe()['75%'],
            df[col].describe()['max']]
    df[col]= pd.cut(df[col], edges, labels = labels, duplicates= 'drop')
    return df
```

```python
labels = ['Not_Popular', 'Below_Avg', 'Averge', 'Popular']
categorize_col(df, 'Vote_Average', labels)
df['Vote_Average'].unique()
```

```python
df.dropna(inplace= True)
df.isna().sum()
```

**We'd spilt Genres into list and then explode our dataframe to have only one genre per row for each movie.**

```python
df['Genre'] = df['Genre'].str.split(', ')
df = df.explode('Genre').reset_index(drop = True)
df.head()
```

```python
#casting column unto category
df['Genre'] = df['Genre'].astype('category')
df['Genre'].dtype
```
